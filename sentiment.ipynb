```python
import pandas as pd
```

```python
import os
import glob
```

```python
dir = './'
dfs = []
for file in os.listdir(dir):
  if "csv" in file:
    print(file)
    data = pd.read_csv(file, index_col=False, delimiter=',', lineterminator='\n')
    dfs.append(data)

all_df = pd.concat(dfs)
all_df
```

```python
all_df = all_df.drop(columns=['Unnamed: 0'],axis=1)
all_df = all_df[all_df['4'].duplicated(keep=False)]
all_df
```

```python
df1 = all_df[all_df['5'].isna()]
df1.drop('5',axis=1, inplace = True)
df1.rename(columns={'0': 'User', '1':'Time', '2':'Likes','3':'Source','4':'Tweet'}, inplace=True)
df1
```

```python
df2 = all_df[~all_df['5'].isna()]
df2.drop('0',axis=1, inplace = True)
df2.rename(columns={'1': 'User', '2':'Time', '3':'Likes','4':'Source','5':'Tweet'}, inplace=True)
df2
```

```python
df1[df1.isna()]
```

```python
dfs = [df1,df2]
all_df = pd.concat(dfs)
all_df.dropna(inplace=True)
all_df = all_df[all_df["Tweet"].str.contains("Attorney Kenneth J. Bush") == False]
all_df
```

```python
all_df['Tweet'].to_list()
```

```python
import transformers
from transformers import pipeline

model_path = 'cardiffnlp/twitter-roberta-base-sentiment-latest'

sentiment_pipeline = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path, device=0)
```

```python

# Using text preprocessing per the model's preprocessing expectations

def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)

all_df['cleaned_tweet'] = all_df['Tweet'].apply(lambda x: preprocess(x))
all_df['cleaned_tweet']
```

```python
from tqdm import tqdm
sentiment_list = []
tweet_list = all_df['cleaned_tweet'].to_list()
for tweet in tqdm(tweet_list):
  analysis = sentiment_pipeline(tweet, max_length=512, truncation=True)
  sentiment_list.append(analysis[0])
```

```python
sentiment_list
```

```python
all_df['sent_analysis'] = [i for i in sentiment_list]
all_df
```

```python

path = 'data/all_tweets_with_sentiment.csv'
with open(path, 'w', encoding = 'utf-8-sig') as f:
  all_df.to_csv(f)
```
